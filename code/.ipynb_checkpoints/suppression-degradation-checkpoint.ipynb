# Interpolation of REBCO layer temperature from Critical Current measured during irradiation

```python
tape, interpolation = 'f20', '0e20pm2'
fitfunction = hts.poly4
path = '/Users/alexisdevitre/Desktop/IcT/{}/{}/'.format(tape, interpolation)
fig, ax = plt.subplots(figsize=(9,4))

fpaths = [path+f for f in os.listdir(path) if (f != '.DS_Store')]
_, _, ic, temperature, _ = hts.getIcT(fpaths, vb=False)

popt, pcov = curve_fit(fitfunction, ic, temperature)
xsmooth = np.linspace(0, 100, 10000)
ysmooth = fitfunction(xsmooth, *popt)
ax.plot(xsmooth, ysmooth, linestyle='-', linewidth=4, alpha=.3, color='b')
ax.plot(ic, temperature, linestyle='None', marker='+', color='k')

ax.set_title('Sample {}, {}\n{:4.2e}x3 + {:4.2e}x2 + {:4.2e}x + {:4.2e}'.format(tape, interpolation, *popt))
ax.set_xlabel('$\mathrm{I_c}$ [A]')
ax.set_ylabel('Temperature [K]')
ax.set_xlim(0, 40)
ax.set_ylim(0, 100)
fig.tight_layout()

print(popt, np.min(ic), np.max(ic))
```
